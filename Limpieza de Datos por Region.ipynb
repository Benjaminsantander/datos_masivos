```python
import pandas as pd
import numpy as np
```

```python
fruta1 = pd.read_json("ordenes_r1.json", typ={'split','index'})
fruta2 = pd.read_json("ordenes_r2.json", typ={'split','index'})

persona_1 = pd.read_json("usuarios_ordenes_r1.json", typ={'split','index'})
persona_2 = pd.read_json("usuarios_ordenes_r2.json", typ={'split','index'})
```

```python
fruta2
```

```python
def trasnformador(lista):
    month = []
    for data in lista:
        value = pd.DataFrame(data)
        value.set_index("orden_id", inplace=True)
        month.append(value)
        
    return month
def trasnformador_persona(lista):
    people = []
    for data in lista:
        value = pd.DataFrame(data)
        people.append(value)
        
    return people
```

```python
frr1 = trasnformador(fruta1)
frr2 = trasnformador(fruta2)

ppr1 = trasnformador_persona(persona_1)
ppr2 = trasnformador_persona(persona_2)
```

```python
frr1[3]
```

```python
ppr1[0]["region"] = 1
ppr1[0]["region"] = 1
```

```python

tabla1 = pd.DataFrame(ppr1[0]["usuario_id"].unique())
tabla2 = pd.DataFrame(ppr2[0]["usuario_id"].unique())
```

```python
tabla1["region"] = 1
tabla2["region"] = 2
```

```python
tabla1
```

```python
result = pd.concat([tabla1, tabla2]).reset_index(drop=True)
result.reset_index(inplace=True)
result.rename(columns={0: "usuario", "index":"id"}, inplace=True)
```
