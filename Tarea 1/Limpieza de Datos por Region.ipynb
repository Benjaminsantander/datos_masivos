```python
def trasnformador(lista):
    month = []
    for data in lista:
        value = pd.DataFrame(data)
        #value.set_index("orden_id", inplace=True)
        value.rename(columns={"limón":"limon", "melón":"melon", "orden_id":"compra_id"}, inplace= True)
        month.append(value)
        
    return month
def trasnformador_persona(lista):
    people = []
    for data in lista:
        value = pd.DataFrame(data)
        people.append(value)
        
    return people
```

```python
frr1 = trasnformador(fruta1)
frr2 = trasnformador(fruta2)

ppr1 = trasnformador_persona(persona_1)
ppr2 = trasnformador_persona(persona_2)
```

```python
def concat_masivo(data):
    value = data[0]
    for pos in range(1, 11):
        value = pd.concat([value, data[pos]])
    value = pd.concat([value, data[11]])
    return value
```

```python
concat_masivo(frr1).columns
```
